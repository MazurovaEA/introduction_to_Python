```python
# 1. Определить корни
import sympy
from sympy import *
from sympy import Symbol
from sympy.solvers import solve
from sympy import oo
x = sympy.Symbol('x',real = True)
f = x**2 - 6*abs(x) + 8
f
```

```python
y = solve(f)
y
```

```python
F=sympy.diff(f, x)
F
```

```python
root_F1 = solveset(F, x, Interval(0, +oo))
root_F1
```

```python
root_F2 = solveset(F, x, Interval(-oo, 0))
root_F2
```

```python
plot(diff(f))
```

```python
intervals_f=[]#интервалы, на которых функция монотонна
intervals_f.append([-oo, int(root_F2.args[0].evalf())])
intervals_f.append([int(root_F2.args[0].evalf()), 0])
intervals_f.append([0, int(root_F1.args[1].evalf())])
intervals_f.append([int(root_F1.args[1].evalf()), +oo])
intervals_f
```

Точка x = 0 является точкой разрыва первого рода

```python
if sympy.is_increasing(f, Interval(intervals_f[0][0], intervals_f[0][1])):
    print(f'На интервале {intervals_f[0]} функция возрастает')
else:
    print(f'На интервале {intervals_f[0]} функция убывает')

if sympy.is_increasing(f, Interval(intervals_f[1][0], intervals_f[1][1])):
    print(f'На интервале {intervals_f[1]} функция возрастает')
else:
    print(f'На интервале {intervals_f[1]} функция убывает')
if sympy.is_increasing(f, Interval(intervals_f[2][0], intervals_f[2][1])):
    print(f'На интервале {intervals_f[2]} функция возрастает')
else:
    print(f'На интервале {intervals_f[2]} функция убывает')
if sympy.is_increasing(f, Interval(intervals_f[3][0], intervals_f[3][1])):
    print(f'На интервале {intervals_f[3]} функция возрастает')
else:
    print(f'На интервале {intervals_f[3]} функция убывает')
```
